```python
from kafka import KafkaConsumer
import matplotlib.pyplot as plt
import json
import pandas as pd
import numpy as np
from PIL import Image
from wordcloud import WordCloud, STOPWORDS
from string import punctuation
from nltk.corpus import stopwords
STOPWORDS=set(stopwords.words('portuguese') + list(punctuation)) 

def create_word_cloud(string):
    maskArray = np.array(Image.open("../assets/cloud.jpg"))
    cloud = WordCloud(background_color = "white", max_font_size=50, max_words = 200, mask = maskArray, stopwords = STOPWORDS)
    cloud.generate(string)
    cloud.to_file("wordCloud.png")
    plt.figure()
    plt.imshow(cloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()
```

```python
brokers = ['localhost:9092']
topico = 'dados-tweets'

consumer = KafkaConsumer(topico, group_id = 'group1', bootstrap_servers = brokers)
```

```python
frases = ''

for messagem in consumer:
    texto = json.loads(messagem.value.decode('utf-8'))
    frases = frases + texto['tweet']
    create_word_cloud(frases)
```
